```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from PIL import Image
import matplotlib.pyplot as plt
import os
try:
   os.makedirs('BinaryFolder')
except OSError:
  pass
#print(os.makedirs('BinaryFolder'))

#Make a folder by the name NewFolder, if anyother name is used then a few strings will have to be changed in the code
#Save all the images in the folder and then run the code

path, dirs, files = next(os.walk("NewFolder"))
file_count = len(files)
#print(file_count)
print (files)

path='/content/NewFolder'

imagearray=[]
for i in range (len(files)):
  extravar = path+'/'+files[i]
 # print (extravar)
  img=Image.open(extravar)
  imagearray.append(img)
  plt.imshow(img)
  #plt.imshow(np.asarray(img), cmap='gray')
  plt.show()
  croppedimg=img.crop((10,510,730,600))
  plt.imshow(croppedimg)
  #plt.imshow(np.asarray(croppedimg), cmap='gray')
  plt.show()
  #Saving the cropped(coloured) image
  newsavingpath='/content/BinaryFolder'+'/cropped'+str (i)
  croppedimg.save(newsavingpath,'png')

  #Converting the Cropped image into binary
  cv2img=cv2.imread('/content/BinaryFolder/cropped'+ str (i),cv2.IMREAD_GRAYSCALE)
  ret, bw_img = cv2.threshold(cv2img, 100, 250, cv2.THRESH_BINARY)
  cv2_imshow(bw_img)
  #Saving the cropped binary image
  cv2.imwrite('/content/BinaryFolder/croppedbinary'+ str (i) +'.png',bw_img)
```
